1. First create a new dataframe **df_by_genre** that contains **genre_1,genre_2,genre_3** and all the columns related to **CVotes/Votes** from the **movies** dataframe. There are 47 columns to be extrected in total.

```python
df_by_genere=movies.loc[:,"CVotes10":"VotesnUS"]
df_by_genere[["genre_1","genre_2","genre_3"]]=movies[["genre_1","genre_2","genre_3"]]
df_by_genere
```

2. Now, Add a column called **cnt** to the dataframe **df_by_genre** and initilize it to one. you will relalise the use of this column by the end of this task.

```python
df_by_genere["cnt"]=1
df_by_genere
```

3. First group the dataframe **df_by_genre** by **genre_1** an dfind the summ of all the numeric columns such as **cnt**, columns related to CVotes and Votes columns and store it in a dataframe **df_by_g1**.

```python
df_by_g_1=df_by_genere.groupby("genre_1").aggregate(np.sum)
df_by_g_1
```

4. Perform the same operation for **genre_2** and **genre_3** and store it dataframes **df_by_g2** and **df_by_g3** respectively.

```python
df_by_g_2=df_by_genere.groupby("genre_2").aggregate(np.sum)
df_by_g_2
```

```python
df_by_g_3=df_by_genere.groupby("genre_3").aggregate(np.sum)
df_by_g_3
```

5.  Now that you have three dataframes performed by grouping over **genre_1,genre_2** and **genre_3** seperately. it's time to combine them. for this, add the three dataframes and store it in new dataframe **df_add**, so that the corrosponding values of Votes/CVotes get added for each genre. there is a function called **add()** in pandas which lets you do this.

```python
df_add=df_by_g_1.add(df_by_g_2,fill_value=0)
df_add
```

```python
df_add=df_add.add(df_by_g_3,fill_value=0)
df_add
```

6. The column **cnt** on aggregation has basically kept the track of the number of occurence of each genre. Subset the genres that have atleast 10 movies into a new dataframe **genre_top10** based on the **cnt** columns value.

```python
genre_top10=df_add[df_add["cnt"]>10]
genre_top10=genre_top10.head(10)
genre_top10
```

7. Now, take the mean of all the numeric columns by dividing them with the column value cnt and store it back to the same dataframe. We will be using this dataframe for further analysis in this task unless it is explicitly mentioned to use the dataframe movies.

```python
genre_top10.iloc[:,0:-1]=genre_top10.iloc[:,0:-1].divide(genre_top10["cnt"],axis=0)
genre_top10
```

8.Since the number of votes can't be a fraction, type cast all the **CVotes** related columns to integers. Also, round off all the Votes related columns upto two digits after the decimal point.

```python
genre_top10.loc[:,"VotesM":"VotesnUS"]=round(genre_top10.loc[:,"VotesM":"VotesnUS"],2)
genre_top10
```

```python
genre_top10[genre_top10.loc[:,"CVotes10":"CVotesnUS"].columns]=genre_top10[genre_top10.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
genre_top10
```
